```r
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import pandas as pd

```

```r
dataset = pd.read_csv('data.csv')
print("Entire Dataset:")
print(dataset)
dataset.head()
```

```r
np.random.seed(0)
X_multi = 2 * np.random.rand(100, 3)
y_multi = 4 + np.dot(X_multi, np.array([3, 2, -1])) + np.random.randn(100)
```

```r
# Define X and y
X = dataset.drop('price', axis=1)
y = dataset['price']  # Target variable
```

```r
# Split the data into training and testing sets
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)

# Initialize the linear regression model
model_multi = LinearRegression()

# Train the model
model_multi.fit(X_train_multi, y_train_multi)

# Make predictions on the test set
y_pred_multi = model_multi.predict(X_test_multi)
```

```r
mse_multi = mean_squared_error(y_test_multi, y_pred_multi)
r2_multi = r2_score(y_test_multi, y_pred_multi)

```

```r
# Plotting actual vs. predicted values for multiple linear regression
plt.figure(figsize=(10, 6))
plt.scatter(y_test_multi, y_pred_multi)
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.title("Actual vs. Predicted values for Multiple Linear Regression")
plt.show()
```

```r
# Print evaluation metrics for multiple linear regression
print("Evaluation metrics for Multiple Linear Regression:")
print("Mean Squared Error:", mse_multi)
print("R-squared:", r2_multi)
```

```r
sns.scatterplot(x='price', y='sqft_above', data=dataset)
plt.title('Scatter plot of Independent vs Dependent Variable')
plt.show()
```

```r
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming y_test and y_pred are 1D arrays (already defined)
# Assuming X_test_multi and y_test_multi were used to obtain y_pred

# Define y_pred using your multiple linear regression model
y_pred = model_multi.predict(X_test_multi)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot: x = Actual Values, y = Predicted Values, z = Data Points
ax.scatter(y_test_multi, y_pred, range(len(y_test_multi)))

# Set labels and title
ax.set_xlabel('Actual Values')
ax.set_ylabel('Predicted Values')
ax.set_zlabel('Data Points')
ax.set_title('3D Plot: Actual vs Predicted Values')

plt.show()

```
