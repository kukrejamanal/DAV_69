# Gauge plot
Gauge charts are also known as dial charts or speedometer charts. This type of chart is often used in executive dashboard reports to show key business indicators. Gauge charts are much useful for comparing the values between a small number of variables either by using multiple needles on the same gauge or by using multiple gauges. It shows the maximum, the minimum, and the present value of data being analyzed. It can be created using the Gauge() method.

```python
import pygal
from IPython.display import SVG, display

# Sample data for demonstration (replace with your actual data)
variable_data = {
    'BMI': 25,
    'Cholesterol': 200,
    'Blood Pressure': 120,
    'Heart Rate': 70
}

# Create a gauge chart
gauge_chart = pygal.Gauge(width=800, height=600)

# Add each variable to the gauge chart with custom color
for i, (variable, value) in enumerate(variable_data.items()):
    # Define custom colors for each variable
    colors = ['#FF5733', '#33FF57', '#337CFF', '#FF33D4']  # You can define more colors if needed
    gauge_chart.add(variable, [{'value': value, 'max_value': 100, 'color': colors[i]}])  # Assuming max value is 100 for demonstration

# Render the chart
display(SVG(gauge_chart.render(disable_xml_declaration=True)))


```

# Funnel chart
Funnel charts are those type of charts which are often used to represent various stages mainly in sales and service process and allows to show the amount of revenue for each of the stages. This chart is much beneficial for identifying potential errors in organization sales and service progress. The funnel chart is similar to the stacked bar chart. The funnel chart helps to summarize data and to track sales conversion and retention rates. And moreover, it can be used to track the success of a business strategy. It can be created using the Funnel() method.

```python
import pygal
from pygal.style import NeonStyle

# Create the chart object
funnel = pygal.Funnel(style=NeonStyle)

# Naming the title
funnel.title = 'Funnel'

# Replace the random data with your actual data from your DataFrame
funnel.add('Age', [26, 22, 39, 39, 32, 30, 33, 24, 24, 30])
funnel.add('Gender', [31, 40, None, None, None, None, 40, 32, 25, 31])
funnel.add('BMI', [37, 27, 31, 20, None, 32, 24, 39, 29, 22])
funnel.add('Chol', [38, None, 20, 29, 33, 23, 32, 33, 32, 23])

# Display the funnel chart
funnel
```

# Treemap in Pygal
Treemap is a chart that is used for displaying echelons data using planted figures mainly rectangles. Each branch of the tree is in rectangular form and then which is macadamize with smaller rectangles to represent sub-branches. The color and the pattern are fixed in such a way in a treemap structure, that it won’t be difficult to understand the material. The second advantage of making treemap is they take less space in layout and can display thousands of items on a screen simultaneously. It can be created using Treemap() method.

```python
# Create a treemap
treemap = pygal.Treemap(style=custom_style)
treemap.title = 'BMI Distribution'

# Add BMI data to the treemap
count = 0
for i, bmi in enumerate(raw['BMI']):
    if(count>2):
      break
    treemap.add(f'Person {i+1}', [{'value': bmi, 'label': f'Person {i+1}'}])
    display(treemap)
    count += 1

```

#Pyramid chart in pygal
Pygal is a Python module that is mainly used to build SVG (Scalar Vector Graphics) graphs and charts. SVG is a vector-based graphics in the XML format that can be edited in any editor. Pygal can create graphs with minimal lines of code that can be easy to understand and write.

```python
import numpy
import pygal

# Generating random data
data = [numpy.random.rand(100), numpy.random.rand(100),
        numpy.random.rand(100), numpy.random.rand(100)]

types = ['A', 'B', 'C', 'D']

# Normalizing data to fit within a specific range, e.g., [0, 1]
normalized_data = [(dat - min(dat)) / (max(dat) - min(dat)) for dat in data]

# Create the pyramid chart
pyramid_chart = pygal.Pyramid(width=400, height=300)  # Adjust width and height as needed

# Naming the title
pyramid_chart.title = 'Pyramid Chart'

# Add normalized data to the chart
for type_, normalized_dat in zip(types, normalized_data):
    pyramid_chart.add(type_, normalized_dat)

display(pyramid_chart)
```

#Radar Chart
The radar chart is a chart or plot that consists of a sequence of equiangular spokes, called radii, with each spoke representing one of the variables. A radar chart is basically a graphical method of displaying data in the form of a two-dimensional chart of three or more quantitative variables that are represented on the axes starting from the same point. Radar charts are MORE helpful for small-to-moderate-sized multivariate data sets. It can be created using the Radar() method.

```python
# Create a radar chart for LDL, Cr, and BUN
radar_chart = pygal.Radar(style=custom_style)
radar_chart.title = 'LDL, Creatinine, and BUN'
radar_chart.x_labels = ['LDL', 'Cr', 'BUN']

# Prepare the data in the correct format
data = [
    {'value': [raw['LDL'][0], raw['Cr'][0], raw['BUN'][0]], 'label': 'Data1'},
    {'value': [raw['LDL'][1], raw['Cr'][1], raw['BUN'][1]], 'label': 'Data2'},
    {'value': [raw['LDL'][2], raw['Cr'][2], raw['BUN'][2]], 'label': 'Data3'},
    {'value': [raw['LDL'][3], raw['Cr'][3], raw['BUN'][3]], 'label': 'Data4'},
    {'value': [raw['LDL'][4], raw['Cr'][4], raw['BUN'][4]], 'label': 'Data5'}
]

# Add the data to the radar chart
for d in data:
    radar_chart.add(d['label'], d['value'])

display(radar_chart)
```
